**Volatility Analysis**

To perform a Volatility Analysis, we will calculate the standard deviation of daily returns for each stock. The standard deviation will serve as a proxy for risk. The most simple definition of volatility is a reflection of the degree to which price moves. A stock with a price that fluctuates wildlyhits new highs and lows or moves erraticallyis considered highly volatile. A stock that maintains a relatively stable price has low volatility.


**Standard deviation**-A standard deviation is a measure of how dispersed the data is in relation to the mean. Low, or small, standard deviation indicates data are clustered tightly around the mean, and high, or large, standard deviation indicates data are more spread out.

```python
standard=df.groupby(['Ticker'])['Closing Price'].std()
df1=pd.DataFrame(standard)
df1.rename(columns={'Closing Price':'Standard Deviation'},inplace=True)
print(df1)
```

```python
df2=pd.DataFrame({'Ticker':['AAPL','GOOG','MSFT','NFLX'],'Standard Deviation':df1['Standard Deviation']})
print(df2)
```

```python
fig2=px.bar(df2,x='Ticker',y='Standard Deviation',color='Ticker',title='Ticker vs Standard Deviation')
fig2.show()
```

From the above bar chart, since **'NFLX'** has the highest standard deviation which implies data for Closing Price of 'NFLX' is closely clustered around the mean closing price of 'NFLX' and that of **'GOOG'** are more spread out.
Also, high deviation means 'NFLX' is more volatile whereas 'GOOG' is less volatile.

Now, the next step is to compare the average values of the Closing prices of the for stocks.



```python
mean1=df.groupby(['Ticker'])['Closing Price'].mean()
df3=pd.DataFrame(mean1)
df3.rename(columns={'Closing Price':'Mean'},inplace=True)
print(df3)
```

```python
df4=pd.DataFrame({'Ticker':['AAPL','GOOG','MSFT','NFLX'],'Mean':df3['Mean']})
print(df4)
```

```python
fig3=px.line(df4,x='Ticker',y='Mean',title='Ticker vs Mean of Closing Price')
fig3.show()
```

The above line chart shows the mean of the closing price of the stocks. Clearly, we can see the **'NFLX'** has the maximum average closing price whereas **'GOOG'** has the minimum.
